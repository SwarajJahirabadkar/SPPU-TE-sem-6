```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.info()
```

```python
plt.bar(data['Age'],data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Fare vs Age Graph")
plt.show()
```

```python
plt.bar(data["Survived"], data["Age"])
plt.xlabel("Survived")
plt.ylabel("Age")
plt.title("Age vs Survival")
```

```python
data.groupby("Sex")['Age'].mean()
```

```python
sns.catplot(x='Sex',y='Age',data = data,kind='box')
plt.show()
```

```python
sns.catplot(x="Parch", y ="Survived", data = data, kind="bar")

plt.show()
```

```python
sns.boxplot(x='Pclass',y='Age',data=data)
```

```python
sns.pairplot(data)
```
